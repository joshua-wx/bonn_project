```python
import os
```

```python
root = '/g/data/kl02/jss548/bonn-project-data/radolan_raw/2017-05'

gunzip_cmd = 'gunzip '
```

```python
#gunzip all files in folder

#recursive
for root, dirs, files in os.walk(root):
    for basename in sorted(files):
        name, ext = os.path.splitext(basename)
        if ext == '.gz':
            #run gunzip
            ffn = os.path.join(root, basename)
            cmd = gunzip_cmd + ffn
            os.system(cmd)
            print('extracted ',basename)
            

print('finished')
            
```
